````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



iris = pd.read_csv(r'./csv_s/Iris - Iris.csv')

setosa = iris[iris['Species'] == 'Iris-setosa']
vesicolor = iris[iris['Species'] == 'Iris-vesicolor']
virginica = iris[iris['Species'] == 'Iris-vriginica']


e = (0, 0.1, 0) # spaces between the categories in the pie chart
avg = list(iris.groupby('Species')['SepalLengthCm'].mean())
plt.pie(avg, explode=e, labels = ['setosa', 'versicolor', 'virginica'])
plt.legend(title = 'Iris Flowers: ')
plt.show()
````

````python

#interpretaion with the box model?
# petal width is very small, and length is also small however larger that width (petal)
# sepal width and length more than that of petal
#the first part the data points are deivated more tna dthe rest haldf not that much



#box plot
setosa.plot.box()
#notch 'True', vert = 0
plt.title('Box Plot')
plt.legend(label = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWitdhCm'])
plt.show()
# the circle are outliers
````
