````python
import pandas as pd
import matplotlib.pyplot as plt


movie = pd.read_csv(r'./csv_s/movies.csv')

animation = movie[movie['Genre'] == 'Animation']
comedy = movie[movie['Genre'] == 'Comedy']
romance = movie[movie['Genre'] == 'Romance']
drama = movie[movie['Genre'] == 'Drama']

#print(movie.groupby('Genre').size())
# print(comedy.describe())

# sepal['SepalWidthCm'].plot(kind = 'hist', label='setosa', color='g', fontsize=10)
# mat.title('Scatter-plot histogram')
# mat.xlabel('SepalWidthCm')
# mat.show()

````

### Scatterplot

````python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.scatter(comedy['Audience score %'], comedy['Profitability'], c='b', marker='o', label='Comedy')
plt.title('Scatter Plot: Audience Score vs. Profitability (Comedy Genre)')
plt.xlabel('Audience Score %')
plt.ylabel('Profitability')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
````

### Bar Graph

````python
genre_avg_scores = movie.groupby('Genre')['Audience score %'].mean()

# Create a bar graph
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
genre_avg_scores.plot(kind='bar', color='b', alpha=0.7)
plt.title('Average Audience Score % by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Audience Score %')

# Show the plot
plt.tight_layout()
plt.show()
````

### Pair plot  using seaborn
Collecting all numeric fiels - 'Audience score %', 'Profitability', 'Rotten Tomatoes %', 'Worldwide Gross'


````python
# Select numeric columns for the pair plot
import searborn as sns

numeric_columns = ['Audience score %', 'Profitability', 'Rotten Tomatoes %', 'Worldwide Gross']

# Create a pair plot
sns.pairplot(movie[numeric_columns], diag_kind='kde')
plt.suptitle('Pair Plot of Numeric Columns', y=1.02)
plt.show()
````

## Advance graphs   

````python


fig, ax = plt.subplots()

fruits = ['Comedy', 'Drama', 'Animation', 'Romance']
counts = [comedy['Worldwide Gross'].describe().iloc[2], 
          romance['Worldwide Gross'].describe().iloc[2],
          drama['Worldwide Gross'].describe().iloc[2], 
          animation['Worldwide Gross'].describe().iloc[2]]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Worldwide Gross')
ax.legend(title='Genre color')

plt.show()
````

### Advance Graph 2

````python
import numpy as np
numeric_columns = ['Audience score %', 'Profitability', 'Rotten Tomatoes %', 'Worldwide Gross']

# Calculate the correlation matrix
correlation_matrix = movie[numeric_columns].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
heatmap = plt.pcolor(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(heatmap)
plt.xticks(np.arange(0.5, len(numeric_columns)), numeric_columns, rotation=45)
plt.yticks(np.arange(0.5, len(numeric_columns)), numeric_columns)
plt.title('Correlation Matrix Heatmap')
plt.show()
````
