```python
import tkinter as tk
from tkinter import messagebox
import re


root = tk.Tk()

root.title('Validation form')
root.geometry('700x500')

headerLabel = tk.Label(root, text='Validation Form', font=('arial', 20)).grid(row=0, columnspan=4)


user_name = tk.StringVar()
nameLabel = tk.Label(root, text='Enter name: ', pady=10, padx=10).grid(row=1, column=0)
nameEntry = tk.Entry(root, textvariable=user_name).grid(row=1, column=1)

user_email = tk.StringVar()
emailLabel = tk.Label(root, text='Enter email:',pady=10).grid(row=2, column=0)
emailEntry = tk.Entry(root,textvariable=user_email ).grid(row=2, column=1)

user_phno = tk.StringVar()
phnoLabel = tk.Label(root, text='Enter ph. no: ', pady=10).grid(row=3, column=0)
phnoEntry = tk.Entry(root, textvariable=user_phno).grid(row=3, column=1)

user_doj = tk.StringVar()
dojLabel = tk.Label(root, text='Enter date of joining (dd/mm/yyy): ', pady=10).grid(row=4, column=0)
dojEntry = tk.Entry(root, textvariable=user_doj).grid(row=4, column=1)

user_dob = tk.StringVar()
dobLabel = tk.Label(root, text='Enter dob (dd/mm/yyy): ',pady=10).grid(row=5, column=0)
dobEntry = tk.Entry(root, textvariable=user_dob).grid(row=5, column=1)

genderLabel = tk.Label(root, text='Select gender: ').grid(row=6, column=0)
radioMale =  tk.Radiobutton(root, text = 'Male', value=1).grid(row=6, column=1)
radioFemale =  tk.Radiobutton(root, text = 'Female', value=2).grid(row=6, column=2)
radioOthers =  tk.Radiobutton(root, text = 'Others', value=3).grid(row=6, column=3)

menuVar = tk.StringVar()
menuVar.set('Branches')
branches = ['Pune', 'Bengluru', 'Dehi', 'Mumbai', 'Kolkata']
branchLable = tk.Label(root, text='Select a branch: ', pady=10).grid(row=7, column=0)
branchMenu = tk.OptionMenu(root, menuVar, *branches).grid(row=7, column=1)

furnitureLabel = tk.Label(root, text='Select furnitures: ').grid(row=8, column=0)
Checkbutton1 = tk.BooleanVar()
Checkbutton2 = tk.BooleanVar()
Checkbutton3 = tk.BooleanVar()
Checkbutton4 = tk.BooleanVar()
Checkbutton5 = tk.BooleanVar()
Checkbutton6 = tk.BooleanVar()

check1 = tk.Checkbutton(root, text = "Sofa", 
                    variable=Checkbutton1,
                      height = 2,
                      width = 10).grid(row=8, column=1)
check2 = tk.Checkbutton(root, text = "Chair", 
                    variable=Checkbutton2,
                      height = 2,
                      width = 10).grid(row=8, column=2)
check3 = tk.Checkbutton(root, text = "Cabinet", 
                    variable=Checkbutton3,
                      height = 2,
                      width = 10).grid(row=8, column=3)
check4 = tk.Checkbutton(root, text = "Curtain", 
                    variable=Checkbutton4,
                      height = 2,
                      width = 10).grid(row=9, column=1)
check5 = tk.Checkbutton(root, text = "Wardrobe", 
                    variable=Checkbutton5,
                      height = 2,
                      width = 10).grid(row=9, column=2)
check6 = tk.Checkbutton(root, text = "Bed", 
                    variable=Checkbutton6,
                      height = 2,
                      pady=10,
                      width = 10).grid(row=9, column=3)


def matches(pattern, my_string) -> bool:
    return len(re.findall(pattern, my_string)) == 1

furniture_dict = {
    "Sofa": 550,
    "Chair": 250,
    "Cabinet": 650,
    "Curtain": 430,
    "Wardrobe": 950,
    "Bed": 600,
}

furniture_list = [550, 250, 650, 430, 950, 600]

states = {
    "check1":False,
    "check2":False,
    "check3":False,
    "check4":False,
    "check5":False,
    "check6":False,
}

def getStates():
    states["check1"] = Checkbutton1.get()
    states["check2"] = Checkbutton2.get()
    states["check3"] = Checkbutton3.get()
    states["check4"] = Checkbutton4.get()
    states["check5"] = Checkbutton5.get()
    states["check6"] = Checkbutton6.get()
 

def registerUser():
    getStates()

    res = {
        "Name": matches(re.compile(r'^[a-zA-Z\s]{3,}$'), user_name.get()),
        "Email": matches(re.compile(r'[a-zA-z0-9.-]+@[a-zA-z]+\.(com|edu|gov)'), user_email.get()),
        "Phno": matches(re.compile(r'[1-9]{1}[0-9]{9}'), user_phno.get()),
        "Date_of_birth": matches(re.compile(r'\d{2}[./\s]\d{2}[./\s]\d{4}'), user_dob.get()),
        "Date_of_joining": matches(re.compile(r'\d{2}[./\s]\d{2}[./\s]\d{4}'), user_doj.get())   
    }

    final = ''
    for field, validated in res.items():
        # TODO fix logic
        if not all(res.values()):
            final += 'Invalid '+res[field]+'\n'
            messagebox.showerror(title="",message=f"Please fill in the fields")
            break
        elif not validated:
            messagebox.showerror(title="",message=f"Please check your {field}")
            break
        elif not any(states.values()):
            messagebox.showerror(title="",message=f"Please select a furniture.")
            break
        else:
            messagebox.showinfo(title="",message=f"You have successfully placed an order")
            break
    result.set(final)
    
        

registerButton = tk.Button(root, text='Register', command=registerUser).grid(row=10, columnspan=4)

result = tk.StringVar()
resultLablel = tk.Label(root, textvariable=result, font=('arial', 12)).grid(row=11, columnspan=4)

furresult = tk.StringVar()
furresultLablel = tk.Label(root, textvariable=result, font=('arial', 12)).grid(row=12, columnspan=4)
root.mainloop()


```
