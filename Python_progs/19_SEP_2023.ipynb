```tk.frame``` - can be a group of widgets <br>
```Text``` - display and alter text <br>
```Label``` - Only display img, text. <br>
```Button``` - display button and handle the functions when clicked <br>
```Entry``` - one line string can be entered into this widget <br>
```Scrollbar``` - this gives scrollbar functionality <br>
```Entry``` - one line string can be entered into this widget <br>

````python
from tkinter import *
from tkinter import messagebox

def register():
    name = name_entry.get()
    email = email_entry.get()
    password = password_entry.get()
    
    print("Name:", name)
    print("Email:", email)
    print("Password:", password)
    print("Gender:", vars.get())
    print("Place of Residence:", cv.get())
    
    name_entry.delete(0, END)
    email_entry.delete(0, END)
    password_entry.delete(0, END)

root = Tk()
root.geometry("500x500")
root.title("Registration Form")

name_label = Label(root, text="Name", font=('arial', 12))
name_label.place(x=60,y=100)
name_entry = Entry(root)
name_entry.place(x=150,y=100)

#NORMAL ENTRY AND LABEL
email_label = Label(root, text="Email", font=('arial', 12))
email_label.place(x=60,y=130)
email_entry = Entry(root)
email_entry.place(x=150,y=130)

#LABEL AND ENTRY WITH DIFFERENT FILL
password_label = Label(root, text="Password", font=('arial', 12))
password_label.place(x=60,y=160)
password_entry = Entry(root, show="*")
password_entry.place(x=150,y=160)

#RADIOBUTTON
gender_label = Label(root, text="Select gender", font=('arial', 12) )
gender_label.place(x=60,y=300)
vars=IntVar()
Radiobutton(root,text="Male",padx=5,variable=vars,value=1,  font=('arial', 12)).place(x=40,y=330)
Radiobutton(root,text="Female",padx=5,variable=vars,value=2,  font=('arial', 12)).place(x=120,y=330)
Radiobutton(root,text="Others",padx=5,variable=vars,value=3,  font=('arial', 12)).place(x=220,y=330)

#DROP LIST
gender_label = Label(root, text="Place of Residence", font=('arial', 12))
gender_label.place(x=60,y=190)
listCountry=("India","Nepal","Other")
cv=StringVar()
drpList=OptionMenu(root,cv,*listCountry)
drpList.config(width=15)
cv.set(listCountry[0])
drpList.place(x=250,y=190)


#CREATE CHECK BOX

cb1 = tk.IntVar()
cb2 = tk.IntVar()
cb3 = tk.IntVar()
cb4 = tk.IntVar()

languaegeLable  = Label(root, text='Languages known: ', font=('Arial', 12)).place(x=90, y=400)

english = tk.Checkbutton(root, text='English', variable=cb1, font=('Arial', 12)).place(x=240, y=400)
tamil = tk.Checkbutton(root, text='Tamil', variable=cb2, font=('Arial', 12)).place(x=240, y=420)
telugu = tk.Checkbutton(root, text='Telugu', variable=cb3, font=('Arial', 12)).place(x=240, y=440)
hindi = tk.Checkbutton(root, text='Hindi', variable=cb4, font=('Arial', 12)).place(x=240, y=460)


#BUTTON
register_button = Button(root, text="Register", command=register,  font=('arial', 12))
register_button.place(x=200,y=430)

# exit_button = Button(root, text="Exit", command=root.destroy())
# exit_button.pack()


# Message.showInfo('Show Info', 'Registeration form is created successfully')


root.mainloop()

````
