````python
import pandas as pd
df = pd.read_csv(r'./csv_s/Iris - Iris.csv')
print(df.columns)

print(df.iloc[1:4])  #iloc stands for integer location 
print(df.head(3))  # both prints the first 3 records


# or a more fancy way of doing this is iterating using for loop:
for index, row in df.iterrows():
    print(index, row['Species'])

print(df.iloc[-1,-1]) # get a specific location (r,c)

print(df.loc) #we sort through the dataset using textual data too, unlike iloc(Int)based
print(df.loc[df['Species'] == 'Iris-virginica'])
````

````python
df.describe()
#sort the values in the dataset
df.sort_values(['SepalWidthCm', 'PetalLengthCm'], ascending=True)

#creating a new column
df['Total'] = df['SepalLengthCm'] + df['SepalWidthCm'] + df['PetalWidthCm']
df.head(4)
df['Total'] = df.iloc[:,1:5].sum(axis=1) #another way of creating a column
#deleting a column
df = df.drop(columns=['Total'])
print(df.head(4))
````
